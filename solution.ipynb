```python
#task 1, importing numpy and pandas 
import numpy as np
import pandas as pd
print("numpy and pandas library installed".upper())
#creating employee data dictionary so that it can be converted to dataframe
#employee dictionary
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}
#converting to dataframe
employee_df=pd.DataFrame(employee_data)
#saving it as .csv file using to_csv()
employee_df.to_csv('employee.csv',index='false')
#display the dataframe
print("\n Employee dataframe created : \n")
print(employee_df)
#creating Seniority Level DataFrame  data dictionary so that it can be converted to dataframe
#Seniority Level dictionary
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 2, 3, 2, 3]
}
#converting to dataframe
seniority_df=pd.DataFrame(seniority_data)
#saving it as .csv file using to_csv()
seniority_df.to_csv('seniority.csv',index='false')
#display the dataframe
print("\n Seniority Level dataframe created : \n")
print(seniority_df)
#creating project data dictionary so that it can be converted to dataframe
#project dictionary ,used np.nan for null values
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, np.nan, 680000, 400000, 350000, np.nan, 300000, 2000000, 1000000, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}
#converting to dataframe
project_df=pd.DataFrame(project_data)
#saving it as .csv file using to_csv()
project_df.to_csv('project.csv',index='false')
#display the dataframe
print("\n project dataframe created : \n")
print(project_df)
print("\n TASK 1 is completed.\n")
```

```python
print("TASK 2")
#checking how many missing values there
print(f"\nMissing values: {project_df['Cost'].isna().sum()}")
#creating a cost list from the project dataframe
costs = project_df['Cost'].tolist()
print('\n',costs)
previous_costs = []
# Loop through the list and update missing values
for i in range(len(costs)):
    if pd.isna(costs[i]):
        # Calculate average of previous non-missing values
        if previous_costs:
            avg = sum(previous_costs) / len(previous_costs)
            costs[i] = avg
    else:
        # Add current value to previous_costs if it's not NaN
        previous_costs.append(costs[i])

# Update the dataframe
project_df['Cost'] = costs

# Optional: Save the cleaned DataFrame
project_df.to_csv("Project.csv", index=False)

# Display result
print('\n',project_df)
print("\n TASK 2 is completed.\n")
```
